```python
import pandas as pd
import numpy as np
```

# Read the CSV and Perform Basic Data Cleaning

```python
df = pd.read_csv("exoplanet_data.csv")
# Drop the null columns where all values are null
df = df.dropna(axis='columns', how='all')
# Drop the null rows
df = df.dropna()
df
```

# Select your features (columns)

```python
# Set features. This will also be used as your x values.
# For this cleaning all error columns were removed. 
selected_features = df
```

# Create a Train Test Split

Use `koi_disposition` for the y values

```python
# Assign the data and the target and pass them as arrays
X = selected_features.drop(columns = ['koi_tce_plnt_num','koi_disposition'])
y = selected_features[['koi_disposition']]
# y = selected_features[['koi_disposition']].ravel()
# y = np.ravel(y)
```

```python
from sklearn.model_selection import train_test_split
X_train, X_test, y_train, y_test = train_test_split(X, y, random_state=2)
# X_train.head()
# type(X_train)
y_train
```

# Pre-processing

Scale the data using the MinMaxScaler and perform some feature selection

```python
# Scale the X data
from sklearn.preprocessing import MinMaxScaler, LabelEncoder
from tensorflow.keras.utils import to_categorical

X_scaler = MinMaxScaler().fit(X_train)
X_train_scaled = X_scaler.transform(X_train)
X_test_scaled = X_scaler.transform(X_test)

print(X.shape, y.shape)
# type(X_test_scaled)
```

```python
# Label encode the responses (y)
label_encoder = LabelEncoder()
label_encoder.fit(y_train)
encoded_y_train = label_encoder.transform(y_train)
encoded_y_test = label_encoder.transform(y_test)
```

```python
# Step 2: Convert encoded labels to one-hot-encoding
y_train_categorical = to_categorical(encoded_y_train)
y_test_categorical = to_categorical(encoded_y_test)

y_test_categorical
y_train_categorical.shape
```

# Construct Neural Network Model


```python
from tensorflow.keras.models import Sequential
from tensorflow.keras.layers import Dense

model = Sequential()
model.add(Dense(units=100, activation='relu', input_dim=39))
model.add(Dense(units=3, activation='softmax'))

# print(f"Training Data Score: {fitted.score(X_train_scaled, y_train)}")
# print(f"Testing Data Score: {fitted.score(X_test_scaled, y_test)}")
```

# Compile the Model

```python
# Compile and fit the model
model.compile(optimizer='adam',
              loss='categorical_crossentropy',
              metrics=['accuracy'])

model.summary()
```

# Fit the Model

```python
model.fit(
    X_train_scaled,
    y_train_categorical,
    epochs=100,
    shuffle=True,
    verbose=2
)
```

```python
# Test the fitted model
# Evaluate the model using the training data
model_loss, model_accuracy = model.evaluate(X_test_scaled, y_test_categorical, verbose=2)
print(f"Loss: {model_loss}, Accuracy: {model_accuracy}")
```

```python
# Add two more layers to the Model, and compile
model2 = Sequential()
model2.add(Dense(units=100, activation='relu', input_dim=39))
model2.add(Dense(units=100, activation='relu'))
model2.add(Dense(units=100, activation='relu'))
model2.add(Dense(units=3, activation='softmax'))
           
model2.compile(optimizer='adam',
              loss='categorical_crossentropy',
              metrics=['accuracy'])
           
model2.summary()
```

```python
# Fit new model
model2.fit(
    X_train_scaled,
    y_train_categorical,
    epochs=100,
    shuffle=True,
    verbose=2
)
```

```python
# Test new model
model_loss, model_accuracy = model2.evaluate(X_test_scaled, y_test_categorical, verbose=2)
print(f"Loss: {model_loss}, Accuracy: {model_accuracy}")
```

# Save the Model

```python
model2.save("JG_Neural_Network.h5")
```
